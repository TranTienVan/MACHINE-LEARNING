```python
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mglearn
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import make_blobs
```

```python
breast_cancer = load_breast_cancer()
print("Breast Cancer keys : {}".format(breast_cancer.keys()))
print("Target names : {}".format(breast_cancer["target_names"]))
print("Feature names : {}".format(breast_cancer["feature_names"]))
print("Target shape : {}".format(breast_cancer["target"].shape))
print("Sample counts per class : \n{}".format({
    n: v for n, v in zip(breast_cancer["target_names"], np.bincount(breast_cancer["target"]))
}))

print("DESCR : {}".format(breast_cancer["DESCR"]))
```

```python
train_accuracy = []
test_accuracy = []

X_train, X_test, y_train, y_test = train_test_split(breast_cancer["data"], breast_cancer["target"], stratify = breast_cancer["target"], random_state = 66)

for n_neighbors in range(1, 11):
    clf = KNeighborsClassifier(n_neighbors = n_neighbors)
    clf.fit(X_train, y_train)
    
    train_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))
    
plt.plot(range(1, 11), train_accuracy, label = "training accuracy")
plt.plot(range(1, 11), test_accuracy, label = "testing accuracy")
plt.legend()
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
```
