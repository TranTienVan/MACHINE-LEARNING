```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
import mglearn
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(random_state = 42)

print("X shape : {}".format(X.shape))
print("y shape : {}".format(y.shape))
print("y : {}".format(y))

mglearn.discrete_scatter(X[:, 0], X[:, 1], y)

plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(["CLass 0", "CLass 1", "CLass 2"])
```

```python
linear_svc = LinearSVC().fit(X, y)
mglearn.discrete_scatter(X[:, 0], X[:, 1],y)
print("Coef shape : {}".format(linear_svc.coef_.shape))
print("Intercept shale : {}".format(linear_svc.intercept_.shape))
line = np.linspace(-15, 15)

for color, coe, inter in zip(["b", "r", "g"], linear_svc.coef_, linear_svc.intercept_):
    plt.plot(line, -(line * coe[0] + inter) / coe[1], color = color)

plt.ylim(-10, 15)
plt.xlim(-10, 8)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(["Class 0", "Class 1", "Class 2", "Line Class 0", "Line Class 1", "Line Class 2"], loc = (1.01, 0.3))

```
